```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x,mlab.normpdf(x, mu, sigma))
plt.show()
```

```python
fig, axes = plt.subplots(2,3, figsize=(10,6), sharex=True, sharey=True)

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
axes[0, 0].plot(x,mlab.normpdf(x, mu, sigma))
axes[0, 0].set_title("Book A")

mu = 1
variance = .5
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
axes[0, 1].plot(x,mlab.normpdf(x, mu, sigma))
axes[0, 1].set_title("Book B")

mu = -.25
variance = 2
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
axes[0, 2].plot(x,mlab.normpdf(x, mu, sigma))
axes[0, 2].set_title("Book C")

mu = 1
variance = 3
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
axes[1, 0].plot(x,mlab.normpdf(x, mu, sigma))
axes[1, 0].set_title("Book D")

mu = -2
variance = 1.75
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
axes[1, 1].plot(x,mlab.normpdf(x, mu, sigma))
axes[1, 1].set_title("Book E")

mu = -.75
variance = .75
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
axes[1, 2].plot(x,mlab.normpdf(x, mu, sigma))
axes[1, 2].set_title("Book F")

fig.tight_layout()
plt.savefig('Six Books.png', dpi=300)
```

```python
fig, axes = plt.subplots(2,3, figsize=(10,6), sharex=True, sharey=True)

thresh = 0.5

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = mlab.normpdf(x, mu, sigma)
axes[0, 0].plot(x, y)
axes[0, 0].fill_between(x[x>thresh], 0, y[x>thresh], facecolor='lightblue')
axes[0, 0].set_title("Book A")

mu = 1
variance = .5
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = mlab.normpdf(x, mu, sigma)
axes[0, 1].plot(x, y)
axes[0, 1].fill_between(x[x>thresh], 0, y[x>thresh], facecolor='lightblue')
axes[0, 1].set_title("Book B")

mu = -.25
variance = 2
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = mlab.normpdf(x, mu, sigma)
axes[0, 2].plot(x, y)
axes[0, 2].fill_between(x[x>thresh], 0, y[x>thresh], facecolor='lightblue')
axes[0, 2].set_title("Book C")

mu = 1
variance = 3
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = mlab.normpdf(x, mu, sigma)
axes[1, 0].plot(x, y)
axes[1, 0].fill_between(x[x>thresh], 0, y[x>thresh], facecolor='lightblue')
axes[1, 0].set_title("Book D")

mu = -2
variance = 1.75
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = mlab.normpdf(x, mu, sigma)
axes[1, 1].plot(x, y)
axes[1, 1].fill_between(x[x>thresh], 0, y[x>thresh], facecolor='lightblue')
axes[1, 1].set_title("Book E")

mu = -.75
variance = .75
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = mlab.normpdf(x, mu, sigma)
axes[1, 2].plot(x, y)
axes[1, 2].fill_between(x[x>thresh], 0, y[x>thresh], facecolor='lightblue')
axes[1, 2].set_title("Book F")

fig.tight_layout()
plt.savefig('Six Books Fill.png', dpi=300)
```
